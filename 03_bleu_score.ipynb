### BLEU score

This notebook demonstrates how to calculate the BLEU score to evaluate the similarity between a generated Persian sentence and a reference sentence.

```
from nltk.translate.bleu_score import sentence_bleu

reference = ["من از این که یک فنجان  قهوه گرم مینوشم خوشحال هستم.".split()]
generated = "من از این که قهوه مینوشم خوشحال هستم.".split()

# Compute BLEU score
bleu_score = sentence_bleu(reference, generated)
bleu_score
```
